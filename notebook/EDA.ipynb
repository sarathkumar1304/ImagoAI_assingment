```python
# Remove outliers using IQR
df.drop('hsi_id',axis=1,inplace=True)
df.drop("vomitoxin_ppb",axis=1)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.decomposition import PCA
```

```python
X = df.iloc[:, :-1]  # First 448 columns
y = df["vomitoxin_ppb"]  # Target variable
```

```python
from matplotlib.pyplot import sca


scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
# Separate Features & Target


corr = df.corr()

# Select Top 10 Correlated Bands with the Target Variable
top_bands = corr["vomitoxin_ppb"].abs().sort_values(ascending=False).index[1:11]

# ---------------- PLOT 1: Histogram of Target Variable ----------------
plt.figure(figsize=(8, 5))
sns.histplot(y, bins=30, kde=True, color="skyblue")
plt.xlabel("Vomitoxin (PPB)")
plt.ylabel("Frequency")
plt.title("Distribution of Vomitoxin (PPB)")
plt.show()
```

### Interpretation:

In summary , the graph shows that most samples have very **low levels** of **Vomitoxin**, but there are few samples with significantly **higher** concentration. This skewed distribution suggest the need for careful monitoring and management of Vomitoxin contamination to ensure food safety

```python
bands = np.arange(X.shape[1])  # Band indices

# ---------------- PLOT 1: Average Reflectance Across Spectral Bands ----------------
mean_reflectance = np.mean(X, axis=0)
std_reflectance = np.std(X, axis=0)

plt.figure(figsize=(12, 6))
plt.plot(bands, mean_reflectance, color="blue", label="Mean Reflectance")
plt.fill_between(bands, mean_reflectance - std_reflectance, mean_reflectance + std_reflectance, color="blue", alpha=0.2, label="Â±1 Std Dev")
plt.xlabel("Spectral Band Index")
plt.ylabel("Reflectance")
plt.title("Average Reflectance Spectrum with Standard Deviation")
plt.legend()
plt.show()
```

### Interpretation:

**Average Reflectance** : The blue line represents the average amount of light reflected at each wavelength.

**Variability in Reflectance** : The shaded blue area around the line indicates the standard deviation, representing the spread or variation in the reflectance measurements.Thus tells us how consistent or inconsistent the reflectance values are.

**Wavelength Range** :The horizontal axis shows the range of wavelength begins measured

**Reflectance Values** : The vertical axis shows the amount of light reflcted ,typically as a ratio or percentage.

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap="coolwarm", alpha=0.7)
plt.colorbar(label="Vomitoxin (PPB)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA of Hyperspectral Data")
plt.show()
```

### Interpretation

The scatter plot represents a <span style="color:orange">Principal Component Analysis (PCA)</span> of hyperspectral data, reducing high dimensional features into two principal components.Each point represents a sample,colored by its <span style="color:orange">vomitoxin (PPB) concentration,</span> with blue indicating lower levels and red indicating higher levels. A few outliers with hih vomitoxin levels (red) are present in the upper left region, suggesting a distanct pattern or anomaly. This visualization helps in <span style="color:orange"> dimenensionlity reduction,clustering and identifying patterns</span> for machine learning application like contamination prediction

```python
import matplotlib.image as mpimg

img = mpimg.imread("actual_vs_predicted.png")  # Load image
plt.imshow(img)  # Display image
plt.axis("off")  # Hide axis
plt.show()
```

#### Key Observations
 - Most points are clustered near the origin, indicating that the model is generally predicting low values.
 - A few scattered points deviate significantly from the red line, suggesting the model has errors and possible underfitting or overfitting issues.
 - Some predictions are notably lower than the actual values, meaning the model might be underestimating higher values.
 - There are extreme outliers where the actual value is high, but the predicted value is much lower, further indicating poor performance in high-value cases.

#### Model Performance Insights
 - If points closely follow the red line: The model is performing well.
 - If points deviate significantly from the red line: The model struggles with predictions, especially for high values.
 - Clustering near the origin but poor prediction for large values: The model might not be well-calibrated for extreme values, possibly due to bias, limited data, or inadequate feature selection.

#### Conclusion
 - The plot suggests that the model is not performing well on high-value predictions, possibly due to underfitting or limited feature representation. Improving the model's predictive ability for high vomitoxin concentrations could lead to better performance.

```python
img = mpimg.imread("residual_plot.png")  
plt.imshow(img)  
plt.axis("off")  
plt.show()
```

#### Key Observations
 - Most residuals are clustered near zero for small predicted values, indicating decent performance in lower ranges.
 - Larger predicted values show higher residuals, meaning the model struggles with high-value predictions.
 - Some residuals are quite large, suggesting that the model has significant errors for certain data points.
 - The spread of residuals is not uniform, which could indicate heteroscedasticity (variance of errors is not constant across predictions).
#### Model Performance Insights
 - If residuals are randomly scattered around zero: The model errors are randomly distributed, which is good.
 - If residuals show a pattern or trend: The model may be biased or failing to capture underlying relationships.
 - If residual variance increases with predicted values: This indicates heteroscedasticity, meaning the model's errors increase as the target value increases, which is a common problem in regression models.

#### Conclusion
The model performs relatively well for lower values but **struggles with high-value predictions**, leading to large residuals and possible **heteroscedasticity**. Improving feature selection, model complexity, or applying transformation techniques can enhance performance.
